```python
df_final['Date_of_Journey'].str.split('/').str[0]
```

```python
## Feature Engineering Process 
## Start from Date_of_Journey

df_final['Date']=df_final['Date_of_Journey'].str.split('/').str[0]
df_final['Month']=df_final['Date_of_Journey'].str.split('/').str[1]
df_final['Year']=df_final['Date_of_Journey'].str.split('/').str[2]
```

```python
df_final.head()
```

```python
## To convert all this new features i.e Date,Month and Year convert into integer
df_final['Date'] = df_final['Date'].astype(int)
df_final['Month'] = df_final['Month'].astype(int)
df_final['Year'] = df_final['Year'].astype(int)
df_final.info()
```

```python
#Drop column or feature

df_final.drop('Date_of_Journey',axis=1,inplace=True)
```

```python
df_final.head(1)
```

```python
## Apply Feature Engineering on Arrival_time

## df_final['Arrival_Time'].str.split(' ').str[0]

## Using Lambda Function

df_final['Arrival_Time']=df_final['Arrival_Time'].apply(lambda x : x.split(' ')[0])
```

```python
df_final.head(1)
```

```python
# To check null value of each features
df_final.isnull().sum()
```

```python
#To obtain arrival_hour and arrival_mintue by feature engineering

df_final['Arrival_hour'] = df_final['Arrival_Time'].str.split(':').str[0]
df_final['Arrival_min'] = df_final['Arrival_Time'].str.split(':').str[1]
```

```python
df_final.head()
```

```python
df_final['Arrival_hour'] = df_final['Arrival_hour'].astype(int)
df_final['Arrival_min'] = df_final['Arrival_min'].astype(int)
```

```python
df_final.info()
```

```python
#Drop arrival_time column
df_final.drop('Arrival_Time',axis = 1,inplace=True)
```

```python
df_final.head(1)
```

```python
## To perform feature engineering on dept_time

df_final['Dept_hour']=df_final['Dep_Time'].str.split(':').str[0]
df_final['Dept_min']=df_final['Dep_Time'].str.split(':').str[1]
```

```python
df_final.head(1)
```

```python
df_final['Dept_hour'] = df_final['Dept_hour'].astype(int)
df_final['Dept_min'] = df_final['Dept_min'].astype(int)
```

```python
df_final.info()
```

```python
# To drop Dept_Time column
df_final.drop('Dep_Time',axis=1,inplace=True)
```

```python
df_final.head(1)
```

```python
df_final['Total_Stops'].unique()
```

```python
# To check null value in total_route with whole obsservation or inform of row

df_final[df_final['Total_Stops'].isnull()]
```

```python
# Now apply feature engineering on Total_Stops column
# Now here we apply map() function to rename that value

df_final['Total_Stops']=df_final['Total_Stops'].map({'non-stop':0,'1 stop':1,'2 stops':2,'3 stops':3,'4 stops':4})
```

```python
df_final.head()
```

```python
# To drop Route feature 
df_final.drop('Route',axis=1,inplace=True)
```

```python
df_final.head(1)
```

```python
df_final['Additional_Info'].nunique()
```

```python
#type(df_final['Duration'].str.split(' ').str[0])

# To search in google as series split in pandas

df_final['Duration_hour']=df_final['Duration'].str.split(' ').str[0].str.split('h').str[0]
```

```python
## Focus on categorical features
df_final['Airline'].unique()
```

```python
# Label Encoding
'''Label Encoding is a technique that is used to convert categorical columns into numerical ones
so that they can be fitted by machine learning models which only take numerical data. 
It is an important pre-processing step in a machine-learning project.'''

from sklearn.preprocessing import LabelEncoder
labelencoder=LabelEncoder()
```

```python
df_final['Airline']=labelencoder.fit_transform(df_final['Airline'])
df_final['Source']=labelencoder.fit_transform(df_final['Source'])
df_final['Destination']=labelencoder.fit_transform(df_final['Destination'])
df_final['Additional_Info']=labelencoder.fit_transform(df_final['Additional_Info'])
```

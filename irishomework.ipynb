```python
# Distribution Plot

#jointplot also used for univariate and bivariate analysis

sns.jointplot(x='sepal_length',y='sepal_width',data=df,kind='hex')
sns.jointplot(x='sepal_length',data=df)
```

```python
sns.jointplot(x='sepal_length',y ='petal_length',data=df,kind='reg')
```

```python
sns.jointplot(x='sepal_length',y ='sepal_width',data=df,kind='reg')
```

```python
#pair plot:- that show pair between each column or one variable of data sets is matched with another variable values

sns.pairplot(df,hue='species')
```

```python
#distplot :- to check distribution for each colunm features

sns.set_style('whitegrid')
sns.distplot(df['sepal_length'],color='red',kde=False,bins=10)
```

```python
#categorical plot

# in this data sets only one categorical colunm is present i.e species

sns.countplot(x='species',data=df)
```

```python
sns.countplot(y='species',data=df)
```

```python
# barplot

sns.barplot(x='species',y='sepal_width',data=df)
```

```python
#box plot :- show the data in five numerical form 
sns.boxplot(x='species',y='sepal_width',data=df,palette='rainbow')
```

```python
# in box plot by default it's take numerical data

# if we doesn't pass x and y axis then it's take all numerical data from data sets

sns.boxplot(data=df,orient='v')
```

```python
#violinplot:-show the data in density and box plot or whisker plot form

sns.violinplot(y='species',x='sepal_length',data=df)
```

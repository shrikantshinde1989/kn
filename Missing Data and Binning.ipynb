## Missing data and Binning

- Missing data Assumptions 
- Missing Completely At Random(MCAR):- Lost Data
- Missing At Random(MAR) :- Missing Only for a specific category
- Missing Not At Random(MNAR) :- Data not collected

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import io

df = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
```

```python
df.head()
#df.info()
df.isnull().sum()
```

```python
#Binning by cut() function:-Bin values into discrete intervals.sort data values into bins
#cut():-useful for going from a continuous variable to a categorical variable.

cut_labels_4 =['kids','young','mid_ages','old']
cut_bins =[0,18,35,45,80]

df['Age_bin'] = pd.cut(df['Age'],bins=cut_bins,labels=cut_labels_4)

df[['Age_bin','Age']]
```

```python
#Uniform Distribution
'''In statistics, uniform distribution refers to a type of probability distribution in which all outcomes are equally 
   likely. A deck of cards has within it uniform distributions because the likelihood of drawing a heart, a club, 
   a diamond, or a spade is equally likely.A coin also has a uniform distribution because the probability of getting 
   either heads or tails in a coin toss is the same. '''
```

```python
#QCut():-Quantile-based discretization function. And give minimum issuse for resultant

bin_labels_5 = ['young','mid aged','old']
df['Age_qcut_bin'] = pd.qcut(df['Age'],q = [0, .4, .8, 1],labels = bin_labels_5)
df[['Age_bin','Age','Age_qcut_bin']]
```

### Univariante Analysis and Univariable
- Uni means one.Hence analysis one variable only.
- Statistically univariate analysis means you have only one outcome variable,but may have multiple independent variables.
  Example:-Linear and Logistic regression
- Unfortunatelyunivariate and univariable terms are used interchangeably.
- Univariable means analyzing one variable only.for example, plots for one variable,
  measures of central tendency(mean,median,mode) and measures of dispersion.
- Histogram,Boxplot,Piechart for univariate analysis
- In univariate analysis,we can not do feature engineering 
- we can do in descriptive statistics 

### Multivariate vs Multivariable
- Multi means many
- Now multivariate and multivariable analyses are used interchangeably
- Statistically multivariate analysis means,you have multiple outcome variables. For eaxmple:-MANOVA,PCA,LDA etc.
- Now multivariate analysis represents multivariable analysis.Means examine many variables.
- Most of used in feature engineering

### Bivariate Analysis
- Bi means two variables.Analysis two variables.For example:- Correlation analysis and scatter plot

```python
# Box plot
### Life Expectency(आयुर्मान)

sns.set_style('whitegrid')
sns.boxplot(data=gapminder_2007,x='continent',y='life_exp')
plt.xlabel('Continent')
plt.ylabel('Life Expectency')
plt.title("Life Expectency of each continent")
```

```python
# Examining the Categorical-Numerical Data using bar plot

#bar plot provide the mean or avrage as estimator='mean'


sns.set_style('darkgrid')
sns.barplot(data=gapminder_2007,x='continent',y='life_exp')
plt.xlabel('Continent')
plt.ylabel('Life Expectency')
plt.title('Life Expectency of each contonent')
```

- box plot and bar plot both can be used to see the difference between categorical and numerical values.
- box plot explain about outlier or skewness.
- bar plot does not explain anything about outlier or skewness.

```python
# showing mean value in box plot

sns.set_style('darkgrid')
sns.boxplot(data=gapminder_2007,x ='continent',y='life_exp',showmeans = True)
plt.xlabel('Continent')
plt.ylabel('Life expectency')
plt.title("Life Expectency of each Continent")
```

#### Examine Numerical-Numerical data to see the relationship

```python
sns.scatterplot(x='gdp_cap',
               y='life_exp',
               data=gapminder_2007)
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title("Relationship Between GDP and Life Expectency")
```

#### we can also add a third variable in scatter plot.hence,it will considered as a multivariable analysis.
- hue:- used for third variable and hue always used categorical feature or column, no need to pass numerical

```python
sns.scatterplot(x='gdp_cap',
               y='life_exp',
               data=gapminder_2007,
               hue='continent')
plt.xlabel('GDP per Capita [in USD]')
plt.ylabel('Life Expectancy [in years]')
plt.title("Relationship Between GDP and Life Expectency")
```

```python
# Converting GDP_Cap column into log scale in scatter plot
#  it is used for log transformation beacuse of scatter plot suggest that data are skewed i.e in right side
# log provide normalization

sns.set_style("whitegrid")
sns.scatterplot(data=gapminder_2007,
               x='gdp_cap',
               y='life_exp')
plt.xscale('log')
plt.xlabel('GDP per Capital in USD')
plt.ylabel('Life Expectency in year')
```

```python
# Converting life expectency into log scale 

sns.set_style('darkgrid')
sns.scatterplot(data=gapminder_2007,
               x='gdp_cap',
               y='life_exp')
plt.xscale('log')#apply log on x-axis and y-axis
plt.yscale('log')
plt.xlabel('GDP Per Capital in USD')
plt.ylabel('Life Expectency in year')
```

### Correlation
- there are two types of the correlation:-1)pearson correlation i.e default method for corr() and 2) spearman correlation
- it always need numerical data
- Pearson correlation evaluate linear relation between two continuous variables.
- Pearson correlation,data must be normal there is no any outlier or skewed data and data in linear fasion
- Spearman correlation does not require or doesn't consider the data is linear or non-linear,they don't care data is normal or not

```python
KC.corr()# default is pearson correlation
```

```python
# by specified method for spearman correlation

KC.corr(method='spearman')
```

```python
# To draw heatmap by pearson corrleation


KC.corr() #selecting the method as pearson
plt.figure(figsize = (20,0)) # setting the figure size
heatmap = sns.heatmap(KC.corr().round(3),vmin=-1,
                     vmax=1,annot=True) # annot = True means writting the data value in each cell.

font2 = {'family':'serif','color':'green','size':20}
plt.title("Pearson Rank Correlation",font2)
```

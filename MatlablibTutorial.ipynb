# MatplotLib Tutirial
- MatplotLib is a plotting library for the python programming language and it's numerical mathematics extension NumPy.
- It provides an object-oriented API for embedding plots into applications using general-purpose GUI toolkits like Tkinter,
    wxPython,QT or GTK+

## Some are major Pros of Matplotlib are:
- Generally easy to get started for simple plots
- Support for custom labels and texts
- Greate control of every elements in a figure
- High-quqlity output in many format
- Very Customizable in genral

```python
import matplotlib.pyplot as plt

%matplotlib inline
import numpy as np
```

```python
# arange() - Return evenly spaced values within a given interval.
x = np.arange(0,10)
y = np.arange(11,21)
```

```python
# plotting using matplotlib

# simple scatter plot

plt.scatter(x,y,c='g')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('Scatter plot in 2D Graph')
```

```python
y = x*x
y
#plt.plot(y)
```

```python
plt.plot(x,y,'b*',linestyle='dashed',linewidth=2,markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2D Diagram')
```

```python
# creating subplot in one canvans i.e in one canvas we create multiple subplot
# creating same plot by different subplot(row,col,color)
plt.subplot(3,3,1)
plt.plot(x,y,'r--')
plt.subplot(3,3,2)
plt.plot(x,y,'b*--')
plt.subplot(3,3,3)
plt.plot(x,y,'ro')
plt.subplot(3,3,4)
plt.plot(x,y,'yX--')
```

```python
# compute x and y coordinates for points sine curve

x = np.arange(0,4*np.pi,0.1)
y = np.sin(x)

# plot the pints using matplotlib
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Sine wave form')
```

```python
# subplot
# compute the x and y coordinates for poins on sine and cosine wave curve
'''plt.subplots_adjust() method to change the space between Matplotlib subplots
    The parameters wspace and hspace specify the space reserved between Matplotlib subplots.
    And the parameters left, right, top and bottom parameters specify four sides of the subplots positions.'''

plt.subplots_adjust(left=0.1,
                    bottom=0.1,
                    right=0.9,
                    top=0.9,
                    wspace=0.4,
                    hspace=0.4)

x = np.arange(0,5 * np.pi,0.1)
y_sine = np.sin(x)
y_cosine = np.cos(x)

#subplot grid that has 2 height and 1 width
# and set the first such subplot as active

plt.subplot(2,1,1)

#make first plot
plt.plot(x,y_sine,'r--')
plt.title('Sine Graph')

# set second subplot as active, and make the second plot
plt.subplot(2,1,2)
plt.plot(x,y_cosine,'b--')
plt.title('Cosine Graph')
```

## Bar Plot
- comapre different list of value
- compare different-differenet bar plot

```python
x1 = [2,8,10]
y1 = [11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]

plt.bar(x1,y1,color = 'y')
plt.bar(x2,y2,color = 'r')
plt.title('Bar Plot')
plt.xlabel('x axis')
plt.ylabel('y axis')
```

## Box plot using matplotlib
- boxplot used to find percentage/percentile
- by using box plot we get outliar

```python
data = [np.random.normal(0,std,100) for std in range(1,4)]

# rectangular Box Plot
plt.boxplot(data,vert = True,patch_artist = True)

# vert = True for vertical boxplot
# patch_artist = True for color inside the box so it's easy to understand for percentile(25%,50%,75%)
```

## Pie chart


```python
# Data to plot in PIE CHART

labels = 'Python','C++','Ruby','Java'
sizes = [215,130,245,210]
explode = (0.2,0,0,0)
colors = ['green','blue','yellow','red']

#plot

plt.pie(sizes,explode = explode,labels = labels,colors = colors,autopct='%1.01f%%',shadow = True)

plt.axis('equal')

          
          
```

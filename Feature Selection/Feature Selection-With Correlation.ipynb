### 2. Feature Selection-With Correlation
- In the step we will be removing the features which are highly correlation

```python
# importing libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
# creating function to find correlation between each features
# with the following fuction we can select highly correlated features
# it will remove the first feature that is correlated with anything other feature


def correlation(dataset,threshold):
    col_corr = set() # set of all the names of correlated columns
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i,j]) > threshold: # we are interesed in absolute coeff value 
                colname = corr_matrix.columns[i] # getting the name of column 
                col_corr.add(colname)
    return col_corr
```

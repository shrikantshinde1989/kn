# Correlation with heatmap
- correlation with heatmap uses colored cells,typically in a monochromatic sccale, to show 2D correlation matrix(table) 
- between two discrete dimensions or event types.it is very important in Feature selection.

```python
# here tip is dependent features
# total bill,smoker,sex,day this are independent features
# correlation find out inter relation between each and every features
# but features must be either int or float value
# correlation not apply on categorical or object type
# value of correlation always in between +1 to -1

```

```python
df.corr()
```

```python
sns.heatmap(df.corr())

# this correlation is important in exploratory data analysis(EDA) when we doing feature selection 
```

# Distribution Plot
- Distplot :- look like histrogram and use for colunm feature to check distribution of colunm feature
- jointplot :- used for univariate and bivariate features to show relation ship between two variables values
- pairplot :- used to paired each variables to another variables and it's used for more than two features 

# jointPlot 
- use to find relationship between two numerical variables. the central chart display their correlation.it is usually a scatterplot, a hexbin plot, 2D histogram or a 2D density plot.
- Univariate Analysis also bivariate 

```python
sns.jointplot(x ='tip',y='total_bill',data = df,kind ='hex')
# kind = 'hex' that show relation in hexgon form
```

```python
sns.jointplot(x = 'tip',y ='total_bill',data = df, kind ='reg')

# it draw kde(kernal density ex) or probability density function(PDF) also draw a regration line or best feet line
```

# PairPlot
- it is also known as scatter plot,in which one variable in same data row is matched with another variable value
- it's showing all variables paired with all the other variables

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='sex')# hue is used to classified the point by different features or categorical features
```

```python
# to count categorical features by each group
df['smoker'].value_counts()
```

```python
sns.pairplot(df,hue = 'smoker')
```

# Dist plot
- Dist plot is help us to check the distribution of the columns feature


```python
sns.set_style('whitegrid')
sns.distplot(df['tip'],color = 'red')

# if KDE is present then it's show the density else where KDE is not present then it's showing count
# by default kde is true,then data in y-axis is showing in percentage
```

```python
sns.distplot(df['total_bill'],color = 'blue',kde = False,bins =10 )
```

# Categorical Plot
- countplot
- boxplot
- barplot
- violinplot

```python
# count plot:- show the count of observation of each categorical bins in form of bar
#              only pass either x-axis or y-axis value we can not pass both axis value at a time
#df['sex'].value_counts()
sns.countplot(x ='sex',data = df)
```

```python
# for y-axis it's count and show in horizontal format

sns.countplot(y='smoker',data=df)
```

```python
# Bar plot

sns.barplot(x='tip',y='sex',data=df)
```

```python
sns.barplot(x='sex',y='total_bill',data = df)
```

# box plot
- box and whisker plot() is a graph that presents information from a five-number summary

```python
sns.boxplot(x='sex',y='total_bill',data=df)
```

```python
sns.boxplot(x ='day',y='total_bill',data=df,palette='rainbow')
```

```python
sns.boxplot(data=df,orient='v')# in box plot by default it's take numerical data

# if we doesn't pass x and y axis then it's take all numerical data from data sets
```

```python
# categorizes data based on some other categorical 

sns.boxplot(x='total_bill',y='day',hue='smoker',data=df)
```

# violinplot
- it's used to see distribution of data in terms of kernal density and box plot

```python
sns.violinplot(x='total_bill',y='day',data=df,hue='smoker',palette='rainbow')
```

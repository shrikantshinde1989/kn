```python
## Concatenation of two different data types variable or addition of integer and string

a=12
b="S"
c=a+b
```

```python
# Occur and handle TypeError Exception

try:
    a=2
    b="s"
    c=a+b
except TypeError:
    print("Try to make the datatype similar")
except NameError:
    print("The user have not defined the variable")
except Exception as ex:
    print(ex)
```

```python
## To handle Zero divided error

try:
    a= int(input("Please Enter the First number:-"))
    b= int(input("Please Enter the Second number:-"))
    c= a/b
    d= a*b
    e= a+b
    print("Division=",c)
    print("Multiplication=",d)
    print("Addition",e)
except NameError:
    print("The user have not defined the variable")
except ZeroDivisionError:
    print("In the division operaion denomrater must be other than zero")
except TypeError:
    print("try to make the datatype must be similar for concatenaton or addition operation")
except Exception as ex:
    print(ex)
```

```python
## Try-Else block 

try:
    a= int(input("Please Enter the first Number:-"))
    b= int(input("Please Enter the Second Number:-"))
    
    c= a/b
    d= a*b
    e= a+b
    
except NameError:
    print("The user have not defined variable")
except ZeroDivisionError:
    print("Please Enter greater than zero nuber(i.e b>0)")
except TypeError:
    print("try to make the datatype must be similar")
except Exception as ex:
    print(ex)

else:
    print(c)
    print(d)
    print(e)
```

```python
## Try-else-finally

try:
    a = int(input("Please Enter the first number:-"))
    b = int(input("Please Enter the second Number:-"))
    c=a/b
except NameError:
    print("The user have not to defined variable")
except TypeError:
    print("Try to make same datatypes")
except ZeroDivisionError:
    print("The number must be grater than zero")
except Exception as ex:
    print(ex)

else:
    print(c)
finally:
    print("Code excution done ")
```

### Custom Exception

```python
class Error(Exception):
    pass

class dobException(Error):
    pass
```

```python
year = int(input("Please Enter the year of birth:-"))
age = 2023-year

print(age)

try:
    if age <= 30 & age >20:
        print("You are eligible for goverment exam")
    else:
        raise dobException
except dobException:
    print("You are not eligible to goverment exam")
```
